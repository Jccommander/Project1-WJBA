```python
#dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#read CSV
edu_df_2016 = pd.read_csv("Output/Edu_Data_2016.csv")
income_df_2016 = pd.read_csv("Output/Income_Data_2016.csv")
mls_attendance = pd.read_csv("Resources/MLS_2017_Attendance.csv")
```

```python
#clean up
edu_df = edu_df_2016.loc[:, ["City & State","Edu Total Pop","Less Than High School Grad","High School Grad","Some College or Associates","Bachelor's Degree or Higher"]]
income_df = income_df_2016.loc[:, ["City & State","Year","Total Households","Less Than 10k","10k to 19k","20k to 34k","35k to 49k","50k to 74k","75k to 99k","100k or More"]]


```

```python
mls_attendance = mls_attendance.dropna()
mls_attendance = mls_attendance.loc[:,["Team","Average"]]
mls_attendance
```

```python
# Create column to hold percentile calculation for edu  group and format it
edu_df["Less Than High School Grad %"] = edu_df["Less Than High School Grad"] / edu_df["Edu Total Pop"] * 100
edu_df["High School Grad %"] = edu_df["High School Grad"] / edu_df["Edu Total Pop"] * 100
edu_df["Some College or Associates %"] = edu_df["Some College or Associates"] / edu_df["Edu Total Pop"] * 100
edu_df["Bachelor's Degree or Higher %"] = edu_df["Bachelor's Degree or Higher"] / edu_df["Edu Total Pop"] * 100
```

```python
# Create column to hold percentile calculation for income  group and format it
income_df["Less Than 10k %"] = income_df["Less Than 10k"] / income_df["Total Households"] * 100
income_df["10k to 19k %"] = income_df["10k to 19k"] / income_df["Total Households"] * 100
income_df["20k to 34k %"] = income_df["20k to 34k"] / income_df["Total Households"] * 100
income_df["35k to 49k %"] = income_df["35k to 49k"] / income_df["Total Households"] * 100
income_df["50k to 74k %"] = income_df["50k to 74k"] / income_df["Total Households"] * 100
income_df["74k to 99k %"] = income_df["75k to 99k"] / income_df["Total Households"] * 100
income_df["100k or More %"] = income_df["100k or More"] / income_df["Total Households"] * 100
```

```python
#edu scatter plot
plt.figure(figsize=(15,10))
plt.scatter(edu_df["Less Than High School Grad %"], mls_attendance["Average"], marker='o',s=100,facecolors='b',edgecolors='black')

plt.title("Attendance vs. Percentage of Population With Less Than High School Education")
plt.ylabel("Average Attendance")
plt.xlabel("Percent of Population, Less Than High School Grad Education")
plt.grid()

plt.show()
```
